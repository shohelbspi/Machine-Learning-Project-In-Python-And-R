```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('president_heights.csv')
```

```python
data.head(2)
```

```python
data.shape
```

```python
data.index
```

```python
hights = np.array(data['height(cm)'])
```

```python
hights
```

```python
print('Mean in Hights : ', hights.mean())
print('Std in Hights : ', hights.std())
print('Max in Hights : ', hights.max())
print('Min in Hights : ', hights.min())
print('Median in Hights : ', np.median(hights))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.hist(hights)

plt.title("Height Distribution of Presidents of USA")
plt.xlabel("height(cm)")
plt.ylabel("Number")
plt.show()
```

```python
plt.plot(hights, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(hights,marker = 'o')
plt.show()
```

```python
plt.bar(data['name'],hights,  width = 0.5)
plt.show()
```

```python
plt.pie(hights, labels = data['name'])
plt.show()
```
